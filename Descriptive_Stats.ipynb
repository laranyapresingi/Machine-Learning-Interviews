```python
"""
Write a function to calculate mean, median, and mode of a list.
Handle edge cases like empty lists and multiple modes.

Example:
Input: [1, 2, 2, 3, 4, 5, 5]
Output: {'mean': 3.14, 'median': 3, 'mode': [2, 5]}
"""

def calculate_central_tendency(data):
    # Your code here
    if not data:
      return {}
    data.sort()
    mean = sum(data) / len(data)
    median = data[len(data) // 2] if len(data) % 2 == 1 else (data[len(data) // 2 - 1] + data[len(data) // 2]) / 2
    mode = max(set(data), key=data.count)
    return {'mean':{mean},'median':{median},'mode':{mode}}
```

```python
calculate_central_tendency([1, 2, 2, 3, 4, 5, 5])
```

Q1. Compute mean, weighted mean, median, and mode from a list of numbers

```python
data = [10, 12, 12, 15, 18, 100]
weights = [1, 1, 2, 1, 1, 1]
import numpy as np
from collections import Counter
data = np.array(data)
weights = np.array(weights)

mean = data.mean()
weighted_mean = np.average(data,weights= weights)
median = np.median(data)

mode = Counter(data).most_common(1)[0][0]

mean, weighted_mean, median, mode
```

Q2. Show why median is more robust than mean to outliers

```python
data_with_outlier = np.append(data, 1000)

np.mean(data), np.mean(data_with_outlier), \
np.median(data), np.median(data_with_outlier)
```

Q3. Implement a trimmed mean

```python
def trimmed_mean(data,trim_ratio = 0.1):
  n = len(data)
  k = int(n * trim_ratio)
  sorted_data = np.sort(data)
  trimmed_data = sorted_data[k:n-k]
  return np.mean(trimmed_data)
trimmed_mean(data_with_outlier)
```
